Create two kernels for convolution: one that looks like an inverted U
and one that looks like a decay function. There is no need to be too
sophisticated in generating, for example, a Gaussian and an exponential;
numerical approximations are fine.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import convolve, square

# 1D inverted U kernel
kernel = np.array([0, 1, 2, 1, 0], dtype=float)
kernel /= kernel.sum()  # normalize

# --- Random Signal ---
N = 100
x_rand = np.random.randn(N)              # random Gaussian signal
y_rand = convolve(x_rand, kernel, mode='same')

# --- Sine Wave ---
t = np.linspace(0, 2*np.pi, 200)
x_sin = np.sin(5*t)
y_sin = convolve(x_sin, kernel, mode='same')

# --- Square Wave ---
x_square = square(5*t)                   # square wave
y_square = convolve(x_square, kernel, mode='same')

# --- Plotting ---
plt.figure(figsize=(10, 8))

plt.subplot(3, 1, 1)
plt.plot(x_rand, label='Original (Random)', color='b')
plt.plot(y_rand, label='Convolved', color='r', linewidth=1.5)
plt.legend(); plt.title('Random Signal with Inverted U Kernel')

plt.subplot(3, 1, 2)
plt.plot(t, x_sin, label='Original (Sine)', color='b')
plt.plot(t, y_sin, label='Convolved', color='r', linewidth=1.5)
plt.legend(); plt.title('Sine Wave with Inverted U Kernel')

plt.subplot(3, 1, 3)
plt.plot(t, x_square, label='Original (Square)', color='b')
plt.plot(t, y_square, label='Convolved', color='r', linewidth=1.5)
plt.legend(); plt.title('Square Wave with Inverted U Kernel')

plt.tight_layout()
plt.show()
```

```python
kernel_decay= np.array([1,0.5,0.25,0.125],dtype=float)
#normalise
kernel_decay /= kernel_decay.sum()

# --- Random Signal with Decay Kernel ---
y_rand_decay = convolve(x_rand, kernel_decay, mode='same')
# --- Sine Wave with Decay Kernel ---
y_sin_decay = convolve(x_sin, kernel_decay, mode='same')
# --- Square Wave with Decay Kernel ---
y_square_decay = convolve(x_square, kernel_decay, mode='same')

# --- Plotting Decay Kernel Results ---
plt.figure(figsize=(10, 8))
plt.subplot(3, 1, 1)
plt.plot(x_rand, label='Original (Random)', color='b')
plt.plot(y_rand_decay, label='Convolved with Decay Kernel', color='r',linewidth=1.5)
plt.legend(); plt.title('Random Signal with Decay Kernel')

plt.subplot(3, 1, 2)
plt.plot(t, x_sin, label='Original (Sine)', color='b')
plt.plot(t, y_sin_decay, label='Convolved with Decay Kernel', color='r',linewidth=1.5)
plt.legend(); plt.title('Sine Wave with Decay Kernel')

plt.subplot(3, 1, 3)
plt.plot(t, x_square, label='Original (Square)', color='b')
plt.plot(t, y_square_decay, label='Convolved with Decay Kernel', color='r',linewidth=1.5)
plt.legend(); plt.title('Square Wave with Decay Kernel')

plt.tight_layout()
plt.show()

         


```
